```python
def display_basic_data_info(df):
    """
    Displays basic information about the dataframe such as head, info, and descriptive statistics.
    """
    print(df.head())
    print()
    print(df.info())
    print()
    print(df.describe())

def plot_stock_price(df):
    """
    Plots the Opening and Closing Prices of the stock.
    """
    plt.figure(figsize=(15, 6))
    df['Open'].plot()
    df['Close'].plot()
    plt.ylabel(None)
    plt.xlabel(None)
    plt.title("Opening & Closing Price of Tesla")
    plt.legend(['Open Price', 'Close Price'])
    plt.tight_layout()
    plt.show()

def plot_stock_volume(df):
    """
    Plots the Sales Volume of the stock.
    """
    plt.figure(figsize=(15, 6))
    df['Volume'].plot()
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title("Sales Volume of Tesla")
    plt.tight_layout()
    plt.show()

def plot_daily_return(df):
    """
    Plots the Daily Return of the stock.
    """
    plt.figure(figsize=(15, 6))
    df['Adj Close'].pct_change().hist(bins=50)
    plt.ylabel('Daily Return')
    plt.title('Tesla Daily Return')
    plt.tight_layout()
    plt.show()

def plot_correlation_matrix(df):
    """
    Plots the correlation matrix of the stock data.
    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title("Correlation Matrix")
    plt.show()

def plot_moving_averages(df):
    """
    Plots moving averages along with the stock's closing price.
    """
    df['MA20'] = df['Close'].rolling(window=20).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    plt.figure(figsize=(15, 6))
    df['Close'].plot(label='Close')
    df['MA20'].plot(label='20-day SMA')
    df['MA50'].plot(label='50-day SMA')
    plt.title("Stock Price with Moving Averages")
    plt.legend()
    plt.show()

def plot_autocorrelation(df):
    """
    Plots the autocorrelation of the stock's closing price.
    """
    autocorrelation_plot(df['Close'])
    plt.title("Autocorrelation of Closing Prices")
    plt.show()
```

## Data Preprocessing

### Choosing Prediction Column

```python
def select_prediction_column(df, column_name):
    """
    Selects the column to be used for prediction and converts it to a dataframe.
    """
    dataset = pd.DataFrame(df[column_name])
    return dataset

# Selecting prediction column
dataset = select_prediction_column(df, "Close")
dataset.shape
```

### Normalizing Data

```python
def normalize_data(data):
    """
    Normalizes the data using MinMaxScaler.
    """
    scaler = MinMaxScaler(feature_range=(0, 1))
    return scaler.fit_transform(np.array(data).reshape(-1, 1))

# Normalizing data
scaled_data = normalize_data(dataset.values)
```

### Splitting the Data

```python
def split_data(scaled_data, train_ratio=0.75):
    """
    Splits the data into training and testing sets.
    """
    train_size = int(len(scaled_data) * train_ratio)
    test_size = len(scaled_data) - train_size
    print("Train Size:", train_size, "Test Size:", test_size)

    train_data = scaled_data[:train_size, 0:1]
    test_data = scaled_data[train_size-60:, 0:1]

    return train_data, test_data, train_size

# Splitting data into training and testing sets
train_data, test_data, train_size = split_data(scaled_data)
```

## Comparison between Models

```python
## Creating Training and Testing Dataset

def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(time_step, len(data)):
        X.append(data[i-time_step:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

time_step = 60
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

## Reshaping Data for RNN Models

X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

## Building RNN Models

def build_simple_rnn(input_shape):
    model = Sequential()
    model.add(SimpleRNN(50, activation='tanh', return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(SimpleRNN(50, activation='tanh', return_sequences=True))
    model.add(Dropout(0.2))
    model.add(SimpleRNN(50, activation='tanh', return_sequences=True))
    model.add(Dropout(0.2))
    model.add(SimpleRNN(50))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(64, return_sequences=False))
    model.add(Dense(32))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

## Training and Evaluating Models

simple_rnn_model = build_simple_rnn((X_train.shape[1], 1))
lstm_model = build_lstm_model((X_train.shape[1], 1))

# Training Simple RNN
history_rnn = simple_rnn_model.fit(X_train, y_train, epochs=30, batch_size=32, verbose=1)

# Training LSTM
history_lstm = lstm_model.fit(X_train, y_train, epochs=30, batch_size=10, verbose=1)
```

```python
## Visualizing Loss

plt.figure(figsize=(10,5))
plt.plot(history_rnn.history['loss'], label='Simple RNN Loss')
plt.plot(history_lstm.history['loss'], label='LSTM Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

# LSTM Model - Enhanced

### Creating training set

```python
def create_dataset(data, time_step=60):
    x, y = [], []
    for i in range(time_step, len(data)):
        x.append(data[i - time_step:i, 0])
        y.append(data[i, 0])
    return np.array(x), np.array(y)

x_train, y_train = create_dataset(train_data)
x_test, y_test = create_dataset(test_data)

# Reshape for LSTM model
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

### Model Structure

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

def build_model(input_shape):
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=input_shape),
        LSTM(64, return_sequences=False),
        Dense(32),
        Dense(16),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=["mean_absolute_error"])
    return model

model = build_model((x_train.shape[1], 1))
model.summary()
```

### Model Training

```python
callbacks = [EarlyStopping(monitor='loss', patience=10, restore_best_weights=True)]
history = model.fit(x_train, y_train, epochs=100, batch_size=32, callbacks=callbacks)
```

### Model Predictions and Evaluation

```python
import matplotlib.pyplot as plt

def plot_history(history):
    plt.plot(history.history["loss"])
    plt.plot(history.history["mean_absolute_error"])
    plt.legend(['Mean Squared Error', 'Mean Absolute Error'])
    plt.title("Losses")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.show()

plot_history(history)
```

```python
def evaluate_model(model, x_test, y_test, scaler):
    predictions = model.predict(x_test)
    predictions = scaler.inverse_transform(predictions)

    y_test_scaled = scaler.inverse_transform([y_test])

    RMSE = np.sqrt(np.mean(y_test_scaled - predictions)**2).round(2)
    print("Root Mean Square Error:", RMSE)

    return predictions

predictions = evaluate_model(model, x_test, y_test, scaler)
```

## Visualizing Predictions

```python
def visualize_predictions(dataset, train_size, predictions):
    train = dataset.iloc[:train_size, :].copy()
    test = dataset.iloc[train_size:, :].copy()

    # Initialize the 'Predictions' column with NaNs
    test['Predictions'] = np.nan

    # Assign predictions using .loc to avoid chained indexing
    test_length = len(test)
    predictions_length = len(predictions)
    test.loc[test.index[-predictions_length:], 'Predictions'] = predictions.flatten()

    plt.figure(figsize=(16, 6))
    plt.title('Stock Price Prediction', fontsize=18)
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Close Price USD ($)', fontsize=18)
    plt.plot(train['Close'], linewidth=3)
    plt.plot(test['Close'], linewidth=3)
    plt.plot(test['Predictions'], linewidth=3)
    plt.legend(['Train', 'Test', 'Predictions'])
    plt.show()

visualize_predictions(dataset, train_size, predictions)
```

## Future Price Prediction

```python
def predict_future_price(data, model, time_step):
    last_sequence = data[-time_step:]
    last_sequence_scaled = scaler.transform(last_sequence.reshape(-1, 1))  # Reshape to 2D array
    last_sequence_scaled = last_sequence_scaled.reshape(1, time_step, 1)   # Reshape to 3D array for prediction
    future_price = model.predict(last_sequence_scaled)
    future_price = scaler.inverse_transform(future_price)
    return future_price[0, 0]

future_price_rnn = predict_future_price(df['Close'].values, simple_rnn_model, time_step)
future_price_lstm = predict_future_price(df['Close'].values, lstm_model, time_step)
print(f"Future Price Prediction with Simple RNN: {future_price_rnn}")
print(f"Future Price Prediction with LSTM: {future_price_lstm}")
```
